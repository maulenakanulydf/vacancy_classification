```python
def preprocess(sentence):
    soup = bs(sentence, features="html.parser")
    sentence = soup.get_text()
    soup = bs(sentence, features="html.parser")
    sentence = soup.get_text()

    sentence = str(sentence)
    sentence = sentence.lower()
    sentence = sentence.replace('{html}',"")
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', sentence)
    rem_url = re.sub(r'http\S+', '',cleantext)
    rem_num = re.sub('[0-9]+', '', rem_url)
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(rem_num)
    filtered_words = [w for w in tokens if not w in stopwords.words('russian')]

    return " ".join(filtered_words)
```

```python
import asyncio
import aiohttp
import pandas as pd
import time

API_URL = "http://51.20.96.227/api/v1/prediction/31ec0563-145a-4701-871a-d57c9dca30ad"
MAX_RETRIES = 3
TIMEOUT = 10  


async def query(session, payload):
    retries = 0
    while retries < MAX_RETRIES:
        try:
            async with session.post(API_URL, json=payload, timeout=TIMEOUT) as response:
                return await response.json()
        except asyncio.TimeoutError:
            retries += 1
            print(f"Timeout occurred. Retrying {retries}/{MAX_RETRIES}...")
            if retries == MAX_RETRIES:
                raise  
            await asyncio.sleep(2)  
        except Exception as e:
            print(f"An error occurred: {e}")
            raise  

async def process_data(df):
    unprocessed_df = df[df['response'].isna() | (df['response'] == "")]
    tasks = []
    
    async with aiohttp.ClientSession() as session:
        for index, row in unprocessed_df.iterrows():
            task = asyncio.create_task(
                process_row(index, row, session, df)
            )
            tasks.append(task)

        await asyncio.gather(*tasks)
        
    df.to_csv('results_with_responses2.csv', index=False)

async def process_row(index, row, session, df):
    try:
        output = await query(session, {"question": row['description']})
        clean_response = output['text'].strip()
        df.at[index, 'response'] = clean_response
        print(f"Processed index {index}: {clean_response}")
    except Exception as e:
        print(f"Failed to process index {index}: {e}")
        df.at[index, 'response'] = f"Error: {str(e)}"

await process_data(df)
```
