```python
class_0_df = df[df['class'] == 67]
```

```python
from collections import Counter
from natasha import Doc, Segmenter, MorphVocab, NewsEmbedding, NewsMorphTagger

# Загружаем необходимые компоненты natasha
segmenter = Segmenter()
morph_vocab = MorphVocab()
emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)
```

```python
all_text = ' '.join(class_0_df['natasha_response_stopwords'].tolist())
```

```python
doc = Doc(all_text)
doc.segment(segmenter)
doc.tag_morph(morph_tagger)
```

```python
lemmas = [] # 67 class
for token in doc.tokens:
    token.lemmatize(morph_vocab)
    lemmas.append(token.lemma)

# Подсчитываем частоту лемм
word_freq = Counter(lemmas)

# Выводим самые частотные леммы
most_common_words = word_freq.most_common(30)
print(most_common_words)
```
